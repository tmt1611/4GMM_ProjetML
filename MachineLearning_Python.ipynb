```python
from math import log, log10

data_transform = data.copy()

#left skewed
data_transform['SM6_L']=np.log((data['SM6_L']+1).max()-data['SM6_L'])
data_transform['SpMax_A']=np.log((data['SpMax_A']+1).max()-data['SpMax_A'])
data_transform['SpMax_L']=np.log((data['SpMax_L']+1).max()-data['SpMax_L'])

#right skewed
data_transform['TI2_L']=np.log(data['TI2_L'])
data_transform['J_Dz(e)']=np.log(data['J_Dz(e)'])
data_transform['Psi_i_A']=np.log(data['Psi_i_A'])
data_transform['SM6_B(m)']=np.log(data['SM6_B(m)'])
data_transform['LOC']=data['LOC'].map(lambda x: np.log10(x+1))
data_transform['SpMax_B(m)']=np.log(data['SpMax_B(m)'])

data_transform['HyWi_B(m)']=np.sqrt(data['HyWi_B(m)'])
```

# 4. Model comparison

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.metrics import f1_score, confusion_matrix, ConfusionMatrixDisplay
def evaluation(model, X_test, Y_test):
    score = model.score(X_test, Y_test)
    print("test error = ",1.-score)
    # Prvision
    y_chap = model.predict(X_test)
    print("F1-score = ", f1_score(Y_test,y_chap))

    plt.rcParams["figure.figsize"] = (3,3)
    cmd = ConfusionMatrixDisplay(confusion_matrix(y_chap, Y_test))
    cmd.plot()
    plt.grid(False)
    plt.show()
```

```python
def plot_feature_coef(coef,names,model_type):

    #Create arrays from feature coef and feature names
    feature_coef = np.array(coef)
    feature_names = np.array(names)

    #Create a DataFrame using a Dictionary
    data={'feature_names':feature_names,'feature_coef':feature_coef}
    fi_df = pd.DataFrame(data)

    #Sort the DataFrame in order decreasing feature coef
    fi_df.sort_values(by=['feature_coef'], ascending=False,inplace=True)

    #Define size of bar plot
    plt.figure(figsize=(10,8))
    #Plot Searborn bar chart
    sns.barplot(x=fi_df['feature_coef'], y=fi_df['feature_names'])
    #Add chart labels
    plt.title(model_type + ' FEATURE COEFFICIENT')
    plt.tick_params(axis='both', which='major', labelsize=20)
    plt.xlabel('FEATURE COEFFICIENT')
    plt.ylabel('FEATURE NAMES')
```

```python
#taken from https://www.analyseup.com/learn-python-for-data-science/python-random-forest-feature-importance-plot.html
def plot_feature_importance(importance,names,model_type):

    #Create arrays from feature importance and feature names
    feature_importance = np.array(importance)
    feature_names = np.array(names)

    #Create a DataFrame using a Dictionary
    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)

    #Sort the DataFrame in order decreasing feature importance
    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)

    #Define size of bar plot
    plt.figure(figsize=(10,8))
    #Plot Searborn bar chart
    sns.barplot(x=fi_df['feature_importance'], y=fi_df['feature_names'])
    #Add chart labels
    plt.title(model_type + ' FEATURE IMPORTANCE')
    plt.tick_params(axis='both', which='major', labelsize=20)
    plt.xlabel('FEATURE IMPORTANCE')
    plt.ylabel('FEATURE NAMES')
    
```
