```python
from math import log, log10

data_transform = data.copy()

#left skewed
data_transform['SM6_L']=np.log((data['SM6_L']+1).max()-data['SM6_L'])
data_transform['SpMax_A']=np.log((data['SpMax_A']+1).max()-data['SpMax_A'])
data_transform['SpMax_L']=np.log((data['SpMax_L']+1).max()-data['SpMax_L'])

#right skewed
data_transform['TI2_L']=np.log(data['TI2_L'])
data_transform['J_Dz(e)']=np.log(data['J_Dz(e)'])
data_transform['Psi_i_A']=np.log(data['Psi_i_A'])
data_transform['SM6_B(m)']=np.log(data['SM6_B(m)'])
data_transform['LOC']=data['LOC'].map(lambda x: np.log10(x+1))
data_transform['SpMax_B(m)']=np.log(data['SpMax_B(m)'])

data_transform['HyWi_B(m)']=np.sqrt(data['HyWi_B(m)'])
```

```python
from sklearn.model_selection import GridSearchCV
import time
```

```python
from sklearn.metrics import confusion_matrix
def evaluation(model, X_test, Y_test):
    score = model.score(X_test, Y_test)
    print("test error = ",1.-score)
    # Prvision
    y_chap = model.predict(X_test)
    
    print('confusion_matrix:')
    table=pd.crosstab(y_chap,Y_test)
    print(table)
```

the SVC implementation in sklearn has O(n) complexity

```python
def ResultSVC (ker):
    param={"C":[0.01, 0.5, 1, 1.4, 2],"gamma":[1, 0.1, 0.01,0.001]}
    svm= GridSearchCV(SVC(kernel=ker,random_state=10),param,cv=10,n_jobs=-1)
    svmOpt=svm.fit(X_train, Y_train)

    # paramtre optimal
    print("Meilleur score = %f, Meilleur paramtre = %s" % (1. - svmOpt.best_score_,svmOpt.best_params_))

    # erreur de prvision sur le test
    print(1-svmOpt.score(X_test,Y_test))

    # prvision de l'chantillon test
    y_chap = svmOpt.predict(X_test)
    # matrice de confusion
    table=pd.crosstab(y_chap,Y_test)
    print(table)
```

```python
for ker in ["linear", "rbf", "sigmoid"]:
    print("result for "+ker+ " kernel")
    ResultSVC(ker)
    print("\n")
```

## Monte-Carlo Cross-Validation

```python
from sklearn.utils import check_random_state
import time
check_random_state(13)
tps0=time.perf_counter()
# dfinition des estimateurs
logit= LogisticRegression(penalty="l1",solver="liblinear")
# tree = DecisionTreeClassifier()
nnet = MLPClassifier(max_iter=600)
# rf   = RandomForestClassifier(n_estimators=100)
svm  = SVC()
# Nombre d'itrations
B = 3 # pour excuter aprs le test, mettre plutt B=30
# dfinition des grilles de paramtres
listMethGrid=[[svm,{"C":[0.4,0.5,0.6,0.8,1,1.4]}],
#    [rf,{"max_features":list(range(2,10,2))}],
    [nnet,{"hidden_layer_sizes":list([(5,),(6,),(7,),(8,)])}],
#    [tree,{"max_depth":list(range(2,10))}],
    [logit,{"C":[0.5,1,5,10,12,15,30]}]]
# Initialisation  0 des erreurs pour chaque mthode (colonne) et chaque itration (ligne)
arrayErreur=np.empty((B,3)) # chua co tree va RF
for i in range(B):   # itrations sur B chantillons test
    # extraction apprentissage et test
    X_train,X_test,Y_train,Y_test=train_test_split(data,biodeg,test_size=200)
    scaler = StandardScaler()  
    scaler.fit(X_train)  
    X_train = scaler.transform(X_train)  
    # Meme transformation sur le test
    X_test = scaler.transform(X_test)
    # optimisation de chaque mthode et calcul de l'erreur sur le test
    for j,(method, grid_list) in enumerate(listMethGrid):
        methodGrid=GridSearchCV(method,grid_list,cv=10,n_jobs=-1).fit(X_train, Y_train)
        methodOpt = methodGrid.best_estimator_
        methFit=methodOpt.fit(X_train, Y_train)
        arrayErreur[i,j]=1-methFit.score(X_test,Y_test)
tps1=time.perf_counter()
print("Temps execution en mn :",(tps1 - tps0))
dataframeErreur=pd.DataFrame(arrayErreur,columns=["SVM","NN","Logit"])   # ,"RF","Tree"
```
