# Training & test set splitting 

## Get dummy variables

```python
quali_list
```

```python
dataDum = pd.get_dummies(data[quali_list])
# del dataDum['F03[C-O]']
del dataDum["B01[C-Br]_0"]
del dataDum["B03[C-Cl]_0"]
dataQuanti = data[quanti_list]
new_data = pd.concat([dataDum,dataQuanti],axis=1)
new_data.head()
```

## Splitting

```python
from sklearn.model_selection import train_test_split  
X_train,X_test,Y_train,Y_test=train_test_split(new_data,biodeg['41'],test_size=200,random_state=11)
```

# Model training

```python
from sklearn.model_selection import GridSearchCV
import time
```

```python
from sklearn.metrics import confusion_matrix
def evaluation(model, X_test, Y_test):
    score = model.score(X_test, Y_test)
    print("test error = ",1.-score)
    # Prvision
    y_chap = model.predict(X_test)
    
    print('confusion_matrix:')
    table=pd.crosstab(y_chap,Y_test)
    print(table)
```

## Logistic regression model with/without variables selection and penalization  

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Optimisation du paramtre de pnalisation
# grille de valeurs
param=[{"C":[1,1.2,1.5,1.7,2,3,4]}]
logit = GridSearchCV(LogisticRegression(penalty="l1",solver="liblinear"), param,cv=5,n_jobs=-1)
logitOpt=logit.fit(X_train, Y_train)  # GridSearchCV est lui-mme un estimateur
# paramtre optimal
logitOpt.best_params_["C"]
print("Meilleur score = %f, Meilleur paramtre = %s" % (1.-logitOpt.best_score_,logitOpt.best_params_))
```

```python
evaluation(logitOpt, X_test, Y_test)
```

```python
# Coefficients
logitLasso=LogisticRegression(penalty="l1",C=logitOpt.best_params_['C'],
                              solver="liblinear")
logitCoef=logitLasso.fit(X_train,Y_train).coef_
print(logitCoef[0])
```

```python
coef = pd.Series(logitCoef[0], index = X_train.columns)
print("Lasso conserve " + str(sum(coef != 0)) + 
      " variables et en supprime " +  str(sum(coef == 0)))
```

```python
imp_coef = coef.sort_values()
plt.rcParams['figure.figsize'] = (6.0, 6.0)
imp_coef.plot(kind = "barh")
plt.title(u"Coefficients du modle lasso")
```

### Linear SVM
We start by implementing linear SVM and see its behavior.

```python
param={"C":[0.01, 0.5, 1, 1.4, 2],"gamma":[1, 0.1, 0.01,0.001], "kernel":["linear", "rbf", "sigmoid"]}
svm= GridSearchCV(SVC(random_state=10),param,cv=10,n_jobs=-1,verbose=True)
svmOpt=svm.fit(X_train, Y_train)

# paramtre optimal
print("Meilleur score = %f, Meilleur paramtre = %s" % (1. - svmOpt.best_score_,svmOpt.best_params_))
print("\n")
# evaluation
evaluation(svmOpt,X_test,Y_test)
```

**Comment**: We did not use "poly" kernel because of too long execution time

the SVC implementation in sklearn has O(n) complexity

```python
def ResultSVC (ker):
    param={"C":[0.01, 0.5, 1, 1.4, 2],"gamma":[1, 0.1, 0.01,0.001]}
    svm= GridSearchCV(SVC(kernel=ker,random_state=10),param,cv=10,n_jobs=-1)
    svmOpt=svm.fit(X_train, Y_train)

    # paramtre optimal
    print("Meilleur score = %f, Meilleur paramtre = %s" % (1. - svmOpt.best_score_,svmOpt.best_params_))

    # erreur de prvision sur le test
    print(1-svmOpt.score(X_test,Y_test))

    # prvision de l'chantillon test
    y_chap = svmOpt.predict(X_test)
    # matrice de confusion
    table=pd.crosstab(y_chap,Y_test)
    print(table)
```

```python
for ker in ["linear", "rbf", "sigmoid"]:
    print("result for "+ker+ " kernel")
    ResultSVC(ker)
    print("\n")
```

## Monte-Carlo Cross-Validation

```python
from sklearn.utils import check_random_state
import time
check_random_state(13)
tps0=time.perf_counter()
# dfinition des estimateurs
logit= LogisticRegression(penalty="l1",solver="liblinear")
# tree = DecisionTreeClassifier()
nnet = MLPClassifier(max_iter=600)
# rf   = RandomForestClassifier(n_estimators=100)
svm  = SVC()
# Nombre d'itrations
B = 3 # pour excuter aprs le test, mettre plutt B=30
# dfinition des grilles de paramtres
listMethGrid=[[svm,{"C":[0.4,0.5,0.6,0.8,1,1.4]}],
#    [rf,{"max_features":list(range(2,10,2))}],
    [nnet,{"hidden_layer_sizes":list([(5,),(6,),(7,),(8,)])}],
#    [tree,{"max_depth":list(range(2,10))}],
    [logit,{"C":[0.5,1,5,10,12,15,30]}]]
# Initialisation  0 des erreurs pour chaque mthode (colonne) et chaque itration (ligne)
arrayErreur=np.empty((B,3)) # chua co tree va RF
for i in range(B):   # itrations sur B chantillons test
    # extraction apprentissage et test
    X_train,X_test,Y_train,Y_test=train_test_split(data,biodeg,test_size=200)
    scaler = StandardScaler()  
    scaler.fit(X_train)  
    X_train = scaler.transform(X_train)  
    # Meme transformation sur le test
    X_test = scaler.transform(X_test)
    # optimisation de chaque mthode et calcul de l'erreur sur le test
    for j,(method, grid_list) in enumerate(listMethGrid):
        methodGrid=GridSearchCV(method,grid_list,cv=10,n_jobs=-1).fit(X_train, Y_train)
        methodOpt = methodGrid.best_estimator_
        methFit=methodOpt.fit(X_train, Y_train)
        arrayErreur[i,j]=1-methFit.score(X_test,Y_test)
tps1=time.perf_counter()
print("Temps execution en mn :",(tps1 - tps0))
dataframeErreur=pd.DataFrame(arrayErreur,columns=["SVM","NN","Logit"])   # ,"RF","Tree"
```

```python
# Distribution des erreurs de prvisions
# Les SVM prsentant des erreurs atypiques sont laisss de ct.
dataframeErreur[["SVM","NN","Logit"]].boxplot(return_type='dict')   #"RF","Tree",
plt.show()
```
